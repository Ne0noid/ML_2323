Этот набор данных позволяет глубоко изучить финансовые показатели Apple за один из самых знаковых периодов в истории компании - с 2009 по 2014 год. В этот период компания выросла из технологического лидера в глобального техногиганта.

Этот набор данных содержит следующие атрибуты:
- EBITDA (millions) – Прибыль до вычета процентов, налогов, износа и амортизации.
- Revenue (millions) – Общая выручка компании.
- Gross Profit (millions) – Валовая прибыль (выручка минус прямые затраты).
- Op Income (millions) – Операционная прибыль (прибыль от основной деятельности).
- Net Income (millions) – Чистая прибыль (после всех расходов и налогов).
- EPS – Прибыль на акцию.
- Shares Outstanding – Количество акций в обращении.
- Year Close Price – Цена закрытия акций на конец года.
- Total Assets (millions) – Общая стоимость активов компании.
- Cash on Hand (millions) – Денежные средства компании.
- Long Term Debt (millions) – Долгосрочные обязательства компании.
- Total Liabilities (millions) – Общие обязательства компании.
- Gross Margin – Валовая маржа (отношение валовой прибыли к выручке).
- PE ratio – Коэффициент цена/прибыль (отношение цены акции к прибыли на акцию).
- Employees – Количество сотрудников компании.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Вывод изначальной таблицы без очистки данных
data_old = pd.read_csv("Apple 2009-2024.csv")
data_old
```

```python
#Загрузка таблицы с очищенными данными (убраны $,в числах запятые заменены на точки)
data = pd.read_csv("Cleaned_data_Apple.csv")
data
```

```python
# Анализ каждого атрибута
columns = data.columns.drop("year")  # Исключаем год из анализа

for col in columns:
    print(f"\n\nАтрибут: {col}")
    print(f"Среднее значение: {data[col].mean():.2f}")
    print(f"Стандартное отклонение: {data[col].std():.2f}")
    
    # Проверка пропущенных значений
    missing_count = data[col].isna().sum()
    print(f"Пропущенные значения: {missing_count}")
    
    # Построение гистограммы
    plt.figure(figsize=(10, 6))
    sns.histplot(data[col], bins=30, kde=True)
    plt.title(f"Гистограмма: {col}")
    plt.xlabel(col)
    plt.ylabel("Частота")
    plt.show()
    
    # Выбросы: межквартильный размах (IQR)
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
        # Нахождение выбросов для текущего столбца
    outliers = data[(data[col] < lower_bound) | (data[col] > upper_bound)]
        
        # Вывод выбросов
    if not outliers.empty:
        print(f"Выбросы для {col}:\n{outliers[['year', col]]}")
    else:
        print(f"Выбросы не найдены для {col}.")
    # Обработка пропущенных значений
    if missing_count > 0:
        print(f"Рекомендация: заполнить пропуски средним ({data[col].mean():.2f}) или медианой ({data[col].median():.2f}).")
```

```python
# Рассчитываем корреляцию между атрибутами
correlation_matrix = data.corr()

# Визуализация корреляционной матрицы
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Матрица корреляции между атрибутами')
plt.show()

# Порог для выявления высококоррелированных атрибутов
threshold = 0.8  # Можно настроить порог

high_corr_pairs = []
for col in correlation_matrix.columns:
    for row in correlation_matrix.index:
        if abs(correlation_matrix.loc[row, col]) > threshold and row != col:
            high_corr_pairs.append((row, col, correlation_matrix.loc[row, col]))

# Выводим высококоррелированные пары
print("Высококоррелированные пары:")
for pair in high_corr_pairs:
    print(f"{pair[0]} и {pair[1]}: {pair[2]:.2f}")

# Построение матрицы графиков рассеивания
sns.pairplot(data)
plt.show()

# Пример выбора некоторых столбцов для построения матрицы графиков рассеивания
selected_columns = ['EBITDA (millions)', 'Revenue (millions)', 'Net Income (millions)', 'Gross Profit (millions)']
sns.pairplot(data[selected_columns])
plt.show()
```
